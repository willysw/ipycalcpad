```python
import numpy as np
import pandas as pd
from pint import UnitRegistry
u = UnitRegistry()

masstable = pd.DataFrame(
    index = ('Main', 'Mezz', 'Roof'),
    columns = ('W_x', 'x'),
    data = ((285.3*u.kip,  9.5*u.ft),
            (220.4*u.kip, 18.5*u.ft),
            (321.9*u.kip, 27.5*u.ft))
)

masstable['Wx'] = masstable['W_x'] * masstable['x']
sum_Wx = masstable['Wx'].sum()

masstable['c_x'] = masstable['Wx'] / sum_Wx.m

masstable
```
